```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pp

epel_dataframe = pd.read_csv('https://data-analysis.fedoraproject.org/csv-reports/countme/totals.csv')
```

```python
df = epel_dataframe
df = df[(df['repo_tag'] == 'epel-8') & (df['week_end'] > '2020-11-25') & (df['os_name'].isin(['Rocky Linux', 'AlmaLinux', 'CentOS Stream', 'CentOS Linux', 'Oracle Linux Server']))]
pt = pd.pivot_table(df, values='hits', index='week_end', columns='os_name', aggfunc=np.sum)
pt.plot(figsize = (24,10))
```

```python
df = epel_dataframe
df = df[(df['repo_tag'] == 'epel-8') & (df['week_end'] > '2021-04-25') & (df['os_name'] == 'Rocky Linux')]
pt = pd.pivot_table(df, values='hits', index='week_end', columns='repo_arch', aggfunc=np.sum)
pt.plot(figsize = (24,10))
```
